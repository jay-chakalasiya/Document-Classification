# TF-IDF

**1)  Term Frequency : TF **: Occurence of word devided by total words in document<br>
 - to remove bias towards the document size<br>
 
**2)  Inverse Document Frequency : IDF **: Logarithm of No of document containing perticular word devided by total number of documents<br>
 - to give weightage towards rearly occuring words<br>

```python
import pandas as pd
import numpy as np
import scipy.spatial.distance as scidis
```

### IDF

```python
df = pd.read_csv("Data/Training Word Count.csv")
df
```

```python
unique_words = list(df)[1:]
Doc_Count=[]
for word in unique_words:
    Current_List = df[word]
    Counter = sum( i>0 for i in Current_List)
    Doc_Count.append(Counter)
```

```python
Doc_Count = np.array(Doc_Count)
Doc_Count = len(df)/Doc_Count
IDF = np.log(Doc_Count)
```

```python
IDF_Frame = pd.DataFrame(unique_words,columns=["Word"])
IDF_Frame["IDF"] = IDF
IDF_Frame.to_csv("Data/IDF.csv")
```

### TF

```python
Document_Labels = df["Label"]
```

```python
df = df.transpose()
df = df[1:]
df
```

```python
ls = list(df.index)
TFIDF_Frame = pd.DataFrame(ls,columns=["Word"])
```

```python
for Document in list(df):
    Current_Array = np.array(df[Document])
    Current_Array = Current_Array*IDF/np.sum(Current_Array)
    
    #----------------------------Normalization---------------------------------------#
    #Max = np.max(Current_Array)
    #Min = np.min(Current_Array)
    #Current_Array = (Current_Array-Min)/(Max-Min)
    
    TFIDF_Frame[Document] = Current_Array
```

```python
TFIDF_Frame = TFIDF_Frame.set_index("Word")
TFIDF_Frame
```

```python
TFIDF_Frame.columns = Document_Labels
TFIDF_Frame
```

```python
TFIDF_Frame = TFIDF_Frame.transpose()
TFIDF_Frame
```

# Outlier Removal

Find distance of all points from origin<br>

Find five number summary

Remove points  **> 1.5*IRQ + Q3**    or      **< Q1-1.5*IRQ**

```python
Origin_Coordinates = np.array([0]*len(unique_words))
len(Origin_Coordinates)
```

```python
Distance_Matrix = []
i=0
while i<len(TFIDF_Frame):
    Current_Document = TFIDF_Frame[i:i+1]
    Current_Document = Current_Document.values.tolist()
    Current_Document = Current_Document[0]
    Current_Document = np.array(Current_Document)
    Distance_Matrix.append(scidis.euclidean(Current_Document,Origin_Coordinates))
    i+=1
Distance_Matrix = np.array(Distance_Matrix)
```

```python
Quartiles = np.percentile(Distance_Matrix, np.arange(0, 100, 25))
Quartiles
```

```python
IRQ = Quartiles[3]-Quartiles[1]
Lower = Quartiles[1]-1.5*IRQ
Upper = Quartiles[3]+1.5*IRQ
```

```python
Final_TF_IDF = pd.DataFrame(unique_words,columns=["Word"])
Labels = list(TFIDF_Frame.index)
New_Labels=[]
```

```python
i=0
while i<len(TFIDF_Frame):
    Current_Document = TFIDF_Frame[i:i+1]
    Current_Document = Current_Document.values.tolist()
    Current_Document = Current_Document[0]
    if Distance_Matrix[i]>Lower and Distance_Matrix[i]<Upper:
        Final_TF_IDF[str(i)]=Current_Document
        New_Labels.append(Labels[i])
    i+=1
```

```python
Final_TF_IDF = Final_TF_IDF.set_index("Word")
Final_TF_IDF.columns = New_Labels
Final_TF_IDF
```

```python
Final_TF_IDF = Final_TF_IDF.transpose()
Final_TF_IDF
```

```python
Final_TF_IDF.to_csv("Data/TF-IDF Training - Normalized.csv")
```
