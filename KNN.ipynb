# KNN on TFIDF

```python
import pandas as pd
import numpy as np
import scipy.spatial.distance as scidis
```

### Normal Data Retrival

```python
Training_Frame = pd.read_csv("Data/TF-IDF Training - Normalized.csv")
Training_Frame = Training_Frame.rename(columns = {'Unnamed: 0':'Label'})
Training_Frame = Training_Frame.set_index("Label")
Training_Frame
```

```python
Testing_Frame = pd.read_csv("Data/TF-IDF Testing - Normalized.csv")
Testing_Frame = Testing_Frame.rename(columns = {'Unnamed: 0':'Label'})
Testing_Frame = Testing_Frame.set_index("Label")
Testing_Frame
```

```python
Training_Data_Labels = list(Training_Frame.index)
Testing_Data_Labels = list(Testing_Frame.index)
len(Training_Data_Labels)
```

```python
Training_Data_Array = np.array(Training_Frame)
Testing_Data_Array = np.array(Testing_Frame)
```

```python
Labels=[]
for Label in Training_Data_Labels:
    if Label not in Labels:
        Labels.append(Label)
```

```python
Weights=[0]*len(Labels)
i=0
while i<len(Weights):
    Weights[i] += Training_Data_Labels.count(Labels[i])
    i+=1
Weights
```

### Normalized Data Retrival

```python
Training_Frame_Normalized = pd.read_csv("Data/TF-IDF Training + Normalized.csv")
Training_Frame_Normalized = Training_Frame_Normalized.rename(columns = {'Unnamed: 0':'Label'})
Training_Frame_Normalized = Training_Frame_Normalized.set_index("Label")
Training_Frame_Normalized
```

```python
Testing_Frame_Normalized = pd.read_csv("Data/TF-IDF Testing + Normalized.csv")
Testing_Frame_Normalized = Testing_Frame_Normalized.rename(columns = {'Unnamed: 0':'Label'})
Testing_Frame_Normalized = Testing_Frame_Normalized.set_index("Label")
Testing_Frame_Normalized
```

```python
Training_Data_Labels_Normalized = list(Training_Frame_Normalized.index)
Testing_Data_Labels_Normalized = list(Testing_Frame_Normalized.index)
len(Training_Data_Labels_Normalized)
```

```python
Training_Data_Array_Normalized = np.array(Training_Frame_Normalized)
Testing_Data_Array_Normalized = np.array(Testing_Frame_Normalized)
```

```python
Labels_Normalized=[]
for Label in Training_Data_Labels_Normalized:
    if Label not in Labels_Normalized:
        Labels_Normalized.append(Label)
```

```python
Weights_Normalized=[0]*len(Labels_Normalized)
i=0
while i<len(Weights_Normalized):
    Weights_Normalized[i] += Training_Data_Labels_Normalized.count(Labels_Normalized[i])
    i+=1
Weights_Normalized
```

## training And Prediction Results With Different Distance Measures

```python
Data_Counter=0
while Data_Counter<2:
    if Data_Counter==0:
        print(" ")
        print(" ")
        print("==========================================   Un-Normalized Data   ============================================")
        Actual_Labels = np.array(list(Testing_Frame.index))
        Current_Testing_Data_Array = np.array(Testing_Data_Array)
        Current_Training_Data_Array = np.array(Training_Data_Array)
        Current_Training_Labels = Training_Data_Labels
        Current_Testing_Labels = Testing_Data_Labels
        Current_Labels = Labels
        Current_Weights = Weights
        
    elif Data_Counter==1:
        print(" ")
        print(" ")
        print("============================================   Normalized Data   ==============================================")
        Actual_Labels = np.array(list(Testing_Frame_Normalized.index))
        Current_Testing_Data_Array = np.array(Testing_Data_Array_Normalized)
        Current_Training_Data_Array = np.array(Training_Data_Array_Normalized)
        Current_Training_Labels = Training_Data_Labels_Normalized
        Current_Testing_Labels = Testing_Data_Labels_Normalized
        Current_Labels = Labels_Normalized 
        Current_Weights = Weights_Normalized
        
    
    
    
    Dis_Counter=0
    while Dis_Counter<4:

        if Dis_Counter==0:
            print(" ")
            print("--------------------Eucledian Distance--------------------------")
            print(" ")
        elif Dis_Counter==1:
            print(" ")
            print("--------------------Minkowski-1 Distance----------------------------------")
            print(" ")
        elif Dis_Counter==2:
            print(" ")
            print("--------------------Minkowski-3 Distance---------------------------")
            print(" ")
        elif Dis_Counter==3:
            print(" ")
            print("---------------------Correlation Distance-----------------------------")
            print(" ")
        k=1
        while k<11:
            Predictions=[]
            Predictions_Weighted=[]
            for Testing_Document in Current_Testing_Data_Array: 

                K_Neighbour_Distance = [float("inf")]*k
                K_Neighbour_Labels = [-1]*k
                Label_Wise_Aggregate_distance = [0]*len(Current_Labels)


                i=0
                for Training_Document in Current_Training_Data_Array:

                    if Dis_Counter==0:
                        Current_Distance = scidis.euclidean(Testing_Document,Training_Document)
                    elif Dis_Counter==1:
                        Current_Distance = scidis.minkowski(Testing_Document,Training_Document,1)
                    elif Dis_Counter==2:
                        Current_Distance = scidis.minkowski(Testing_Document,Training_Document,3)
                    elif Dis_Counter==3:
                        Current_Distance = scidis.correlation(Testing_Document,Training_Document)




                    Maximum_Valued_Index = np.argmax(K_Neighbour_Distance)

                    if K_Neighbour_Distance[Maximum_Valued_Index] > Current_Distance:
                        K_Neighbour_Distance[Maximum_Valued_Index] = Current_Distance
                        K_Neighbour_Labels[Maximum_Valued_Index] = Current_Training_Labels[i]

                    j=0
                    for Label in Current_Labels:
                        l=0
                        for K_Neighbour_Label in K_Neighbour_Labels:
                            if K_Neighbour_Label==Label:
                                if K_Neighbour_Distance[l]==0.0:
                                    Label_Wise_Aggregate_distance[j]+=float("inf")
                                else:
                                    Label_Wise_Aggregate_distance[j]+=1/K_Neighbour_Distance[l]
                            l+=1
                        j+=1

                    i+=1




                Label_Wise_Aggregate_distance = np.array(Label_Wise_Aggregate_distance)
                Most_Nearer_Label_Index = np.argmax(Label_Wise_Aggregate_distance)

                Label_Wise_Aggregate_weighted_distance = np.array(Label_Wise_Aggregate_distance/np.array(Current_Weights))
                Most_Nearer_Weighted_Label_Index = np.argmax(Label_Wise_Aggregate_weighted_distance)

                Predictions.append(Current_Labels[Most_Nearer_Label_Index])
                Predictions_Weighted.append(Current_Labels[Most_Nearer_Weighted_Label_Index])

            Predictions = np.array(Predictions)
            Predictions_Weighted = np.array(Predictions_Weighted)

            Truths = list((Actual_Labels==Predictions)).count(True)
            Truths_Weighted = list((Actual_Labels==Predictions_Weighted)).count(True)

            print("K = "+ str(k) +" ------ Efficiency : "+str(int(Truths*100/len(Actual_Labels)))+ "% ----------- With Weights : "+str(int(Truths_Weighted*100/len(Actual_Labels)))+ "%")

            k+=1
        Dis_Counter+=1
    Data_Counter+=1
```
